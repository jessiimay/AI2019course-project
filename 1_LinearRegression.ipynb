```python
#å¤„ç†ç¼ºå¤±å€¼
df=pd.read_excel(io=IO)
means = df.replace(-200,0).mean(0).values
df
```

```python
#ç¼ºå¤±å€¼å¤„ç†
for i in range (2,df.shape[1]):
    for j in range (0,df.shape[0]):
        if(df.iloc[j,i]==-200):
            df.iloc[j,i]=means[i-2]
```

```python
X=df.iloc[:,2:14]
y=df.iloc[:,14]
print(X,y)
```
